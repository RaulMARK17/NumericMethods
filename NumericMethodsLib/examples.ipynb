# Ecuaciones diferenciales

```python
e = nm.euler("-0.21972245773*(y - 25)", 0, 41, xn=25, h=0.05, n=10)
```

```python
import matplotlib.pyplot as plt

def ED(x, y):
    return -0.21972245773*(y - 25)

n = 500
x0 = 0
y0 = 41
h = 0.05
y_ = [y0]
x = [x0]

for i in range(n):
    yi = y_[i] + ED(x[i], y_[i])*h
    y_.append(yi)
    x.append(x[i] + h)
print('x = {}, y = {}'.format(x[-1], y_[-1]))

plt.plot(x, y_)
plt.show()    
```

```python
e.compare('25 + 75*np.exp(-0.21972245773*np.array(x))')
```

```python
import numpy as np
def f(x):
    return 25 + 75*np.exp(-0.21972245773*np.array(x))

plt.plot(x, y_)
plt.plot(x, f(x), c='red')
plt.show()

print('Error = {}%'.format((f(x[-1]) - y_[-1])/f(x[-1])))
```

```python
ime = nm.eulerImproved("-0.21972245773*(y - 25)", 0, 41, xn=25, h=0.05, n=10)
```

```python
import matplotlib.pyplot as plt

def ED(x, y):
    return -0.21972245773*(y - 25)

n = 500
x0 = 0
y0 = 41
h = 0.05
y_ = [y0]
x = [x0]

for i in range(0, n):
    x.append(x[i] + h)
    yi_ = y_[i] + h*ED(x[i], y_[i])
    yi = y_[i] + h*(ED(x[i], y_[i]) + ED(x[i+1], yi_))/2
    y_.append(yi)
    
print('x = {}, y = {}'.format(x[-1], y_[-1])) 

plt.plot(x, y_)
plt.show()  
```

```python
ime.compare('25 + 75*np.exp(-0.21972245773*np.array(x))')
```

```python
def f(x):
    return 25 + 75*np.exp(-0.21972245773*np.array(x))

plt.plot(x, y_)
plt.plot(x, f(x), c='red')
plt.show()

print('Error = {}%'.format((f(x[-1]) - y_[-1])/f(x[-1])))
```

```python
rk = nm.rungeKutta_4("-0.21972245773*(y - 25)", 0, 100, xn=25, h=0.05, n=10)
```

```python
import matplotlib.pyplot as plt

def ED(x, y):
    return -0.21972245773*(y - 25)
    
n = 500
x = [0]
y_ = [100]
h = 0.05

for i in range(n):
    k1 = h*ED(x[i], y_[i])
    k2 = h*ED(x[i] + 0.5*h, y_[i] + 0.5*k1)
    k3 = h*ED(x[i] + 0.5*h, y_[i] + 0.5*k2)
    k4 = h*ED(x[i] + h, y_[i] + k3)

    y_.append(y_[i] + (1/6)*(k1 + 2*k2 + 2*k3 + k4))
    x.append(x[i] + h)


plt.plot(x, y_)
plt.show()
            
```

```python
rk.compare('25 + 75*np.exp(-0.21972245773*np.array(x))')
```

```python
def f(x):
    return 25 + 75*np.exp(-0.21972245773*np.array(x))

plt.plot(x, y_)
plt.plot(x, f(x), c='red')
plt.show()

print('Error = {}%'.format((f(x[-1]) - y_[-1])/f(x[-1])))
```

# NewtonRapshon

```python
nr = nm.newtonRaphson(10, "x**2 - 10*x + 15")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = 10
limit = 10000
p = 1e-05

def f(x):
    return x**2 - 10*x + 15

def derivative(x):
    h = 0.000001
    return (-f(x + 2*h) + 8*f(x+h) - 8*f(x-h) + f(x - 2*h))/(12*h)

iteration = 0
while((f(x) > p) or (f(x) < -p) or (x < 0)):
    xi = (x - f(x)/derivative(x))
    x = xi
    iteration += 1
    if iteration >= limit:
        break

print('x = {}'.format(x))
print('f(x) = {}'.format(f(x)))
print('iteración: {}'.format(iteration))

X = np.linspace(-1+x, x+1, 100)

plt.plot(X, f(X))
plt.plot(x, f(x), 'o', c='red')
plt.show()
```

# Cálculo

```python
c = nm.calculus()
```

```python
c.derive("x**2", 4, 0.000001)
```

```python
x = 4
h = 1e-06
def f(x):
    return x**2

def derive(x, h):
    return (-f(x + 2*h) + 8*f(x+h) - 8*f(x-h) + f(x - 2*h))/(12*h)

print(derive(x, h))
```

```python
c.integrate('x**2', 0, 10)
```

```python
a = 0
b = 10
n = 10000

def f(x):
    return x**2

def integrate(a, b, n):
    _ = (b-a)/n
    increase = a
    integral = 0
    for i in range(n):
        integral += (f(increase) + f(increase + _))*((increase + _) - increase)/2
        increase += _
    return integral

print(integrate(a, b, n))
```
